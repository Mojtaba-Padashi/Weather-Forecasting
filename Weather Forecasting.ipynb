```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from prophet import Prophet
from prophet.plot import plot_plotly , plot_components_plotly
```

```python
data = pd.read_csv('DailyDelhiclimateTrain.csv')
print(data.head())
```

```python
print(data.describe())
```

```python
print(data.info())
```

```python
figure = px.line(data , x = 'date' , y = 'meantemp' , title = 'Mean Temperature in Delhi Over The Years')
figure.show()
```

```python
figure = px.line(data , x = 'date' , y = 'humidity' , title = 'Humidity in Delhi Over The Years ')
figure.show()
```

```python
figure = px.line(data , x = 'date' , y = 'wind_speed' , title = 'Wind Speed in Delhi Over The Years')
figure.show()
```

```python
figure = px.scatter(data_frame = data , x = 'humidity' , y = 'meantemp' , size = 'meantemp' , trendline = 'ols' , title = 'Relationship Between Temperature and Huimidity')
figure.show()
```

```python
data['date'] = pd.to_datetime(data['date'] , format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
print(data.head())
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figure = (15 , 10))
plt.title('Temperature Change in Delhi Over The Years')
sns.lineplot(data = data , x = 'month' , y = 'meantemp' , hue = 'year')
plt.show()
```

```python
forecast_data = data.rename(columns = {'date':'ds' , 'meantemp':'y'})
print(forecast_data)
```

```python
model = Prophet()
model.fit(forecast_data)
forcast = model.make_future_dataframe(periods = 365)
predictions = model.predict(forcast)
plot_plotly(model , predictions)
```
